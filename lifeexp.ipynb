```python
from scipy import  stats
```

```python

def pearson_test(df, num, target='Life expectancy '):
    for i in num:
        x_num = df[i]
        y_num = df[target]
        r, p = stats.pearsonr(x_num, y_num)
        print(f'Pearson correlation between {i} and {target}: r={r:.4f}, p-value={p:.4f}')

# Example usage
pearson_test(df, num)
```
